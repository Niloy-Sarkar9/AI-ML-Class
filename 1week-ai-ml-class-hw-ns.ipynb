```python
# 5 Python Program to Print Sum of Negative Numbers, Positive Even Numbers and Positive Odd numbers in a List
lst = [-4,-5,8,-11,14,18,-19,22,25,-26]
z=[i for i in lst if i<0]
x=[i for i in lst if i>0 and i%2==0]
y=[i for i in lst if i>0 and i%2!=0]
print("negative numbers sum",sum(z))
print("even positive numbers sum",sum(x))
print("odd positive numbers sum",sum(y))
```

```python
# 8 Python Program to Print Largest Even and Largest Odd Number in a List
def printmax(l):
    l = [11,54,83,42,98,421,153,654,246]
    even = [x for x in l if x % 2 == 0]
    odd = [x for x in l if x % 2 == 1]
    print("Largest even number is ", max(even))
    print("Largest odd number is ", max(odd))
printmax(l)
```
