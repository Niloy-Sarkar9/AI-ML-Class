```python
#1 Create a 5X2 integer array and Prints its attributes
import numpy as np
array = np.arange(100, 200, 10).reshape(5, 2)
print(array)
print("Array Shape is:", array.shape)
print("Array dimensions are:", array.ndim)
print("Length of each element of array in bytes is:", array.itemsize)
```

```python
#2 Create an array by adding the following two NumPy arrays
import numpy as np

arrayOne = np.array([[5, 6, 9], [21, 18, 27]])
arrayTwo = np.array([[15, 33, 24], [4, 7, 1]])
result = arrayOne + arrayTwo
squared_result = np.square(result)
print(result)
print(squared_result)
```

```python
#3 Find the length of each string element in the Numpy array 
import numpy as np
arr = np.array(['Python', 'Week3', 'home', 'work'])
sizes = [len(i) for i in arr]
print(sizes)
```

```python
#4 Find Mean ,median and mode of a List of Numpy Array
import numpy as np
array = [4,5,8,11,15,24,29,33,42,46,50]
mean = np.mean(array)
median = np.median(array)
mode = np.unique(array)[np.argmax(np.bincount(array))]
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode}")
```

```python
#5 .Calculate the Euclidean distance using NumPy
import numpy as np
a = np.array([2, 4, 8])
b = np.array([3, 5, 10])

# Calculate the Euclidean distance
dist = np.linalg.norm(a - b)
print(f"Euclidean Distance: {dist:.2f}")
```

```python
#6 COVID-19 Data Visualization using matplotlib in Python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create random example COVID-19 dataset
dates = pd.date_range(start='2021-01-01', periods=10, freq='D')
daily_confirmed = np.random.randint(low=25, high=75, size=10)

covid_data = pd.DataFrame({'Date': dates, 'Daily Confirmed': daily_confirmed})

plt.plot(covid_data['Date'], covid_data['Daily Confirmed'], marker='*', linestyle='--', color='r')
plt.title("COVID-19 Daily Confirmed Cases")
plt.show()
```
